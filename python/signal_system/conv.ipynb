```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
```

# 入门卷积

```python
x = np.array([1,1,1,1])
h = np.power(1.2,[0,1,2,3,4,5,6])
y = signal.convolve(x, h)
```

```python
fig, axes = plt.subplots(3, 1, sharex=True)
axes[0].stem(x)
axes[1].stem(h)
axes[2].stem(y)
```

# 入门卷积 + 采样

```python
x = np.repeat([1,1,1,1], 100)
h = np.power(1.2,np.arange(1,8, 1/400))
y = signal.convolve(x, h)
```

```python
fig, axes = plt.subplots(3, 1, sharex=True)
axes[0].plot(x)
axes[1].plot(h)
axes[2].plot(y)
```
